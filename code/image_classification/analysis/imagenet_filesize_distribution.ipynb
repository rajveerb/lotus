```python
import os
# plot the distribution of JPEG file sizes with matplotlib with interval size 100
import matplotlib.pyplot as plt
```

```python

def get_jpeg_sizes(path):
    sizes = []
    for dirpath, dirnames, filenames in os.walk(path):
        for filename in filenames:
            if filename.lower().endswith('.jpeg'):
                filepath = os.path.join(dirpath, filename)
                sizes.append(os.path.getsize(filepath))
    return sizes

path = '/home/hice1/mrao70/scratch/special_problems/tiny-imagenet-200/train'
jpeg_sizes = get_jpeg_sizes(path)
print(f'Sizes of JPEG files in {path}: {len(jpeg_sizes)}')
```

```python
# sort the sizes
jpeg_sizes.sort()
```

```python
print(f"min file size and max file size: {min(jpeg_sizes)} and {max(jpeg_sizes)}")
print(f"gap size: {(max(jpeg_sizes) - min(jpeg_sizes))/len(jpeg_sizes)}")
```

```python
# add lo scale on x-axis
plt.xscale('log')
plt.hist(jpeg_sizes, bins=range(min(jpeg_sizes),max(jpeg_sizes), 100))
# make x axis and y axis labels larger
plt.xlabel('JPEG file size (bytes)', fontsize=20)
plt.ylabel('Number of files', fontsize=20)
# make labels larger
plt.tick_params(axis='both', which='major', labelsize=16)


plt.show()
```

```python
# 99% of the files are less than 370KB in size
print(f"99% of the files are less than {jpeg_sizes[int(len(jpeg_sizes)*0.99)]/1024} KB in size")
```

```python
# get median file size
print(f"median file size: {jpeg_sizes[int(len(jpeg_sizes)*0.5)]/1024:2f} KB")
# get mean
print(f"mean file size: {sum(jpeg_sizes)/len(jpeg_sizes)/1024:2f} KB")
# get P99
print(f"P99 file size: {jpeg_sizes[int(len(jpeg_sizes)*0.99)]/1024:2f} KB")
# get P75
print(f"P75 file size: {jpeg_sizes[int(len(jpeg_sizes)*0.75)]/1024:2f} KB")
# get P25
print(f"P25 file size: {jpeg_sizes[int(len(jpeg_sizes)*0.25)]/1024:2f} KB")
# total data size
print(f"total data size: {sum(jpeg_sizes)/1024/1024/1024:2f} GB")
# get standard deviation using numpy
import numpy as np
print(f"standard deviation: {np.std(jpeg_sizes)/1024:.2f} KB")
```
